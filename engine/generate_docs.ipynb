## 1. Diagram Helper Functions
We define custom helpers using standard `matplotlib` to draw crisp, professional, vector-style block diagrams, class structures, use cases, and entity-relationship models without needing external Graphviz software.

```python
def draw_box(ax, x, y, title, content, bg_color, width=2.4, height=1.2, title_color='#2c3e50'):
    """Draws a rounded box with a header and main body text."""
    # Draw shadow first
    shadow = patches.FancyBboxPatch(
        (x - width/2 + 0.05, y - height/2 - 0.05),
        width, height,
        boxstyle="round,pad=0.1",
        facecolor='#d5dbdb', edgecolor='none', zorder=1
    )
    ax.add_patch(shadow)
    
    # Draw main box
    box = patches.FancyBboxPatch(
        (x - width/2, y - height/2),
        width, height,
        boxstyle="round,pad=0.1",
        facecolor=bg_color, edgecolor='#34495e', lw=1.5, zorder=2
    )
    ax.add_patch(box)
    
    # Add text
    ax.text(x, y + height*0.25, title, ha='center', va='center', fontsize=10, weight='bold', color=title_color, zorder=3)
    ax.text(x, y - height*0.15, content, ha='center', va='center', fontsize=8, color='#2c3e50', zorder=3)

def draw_arrow(ax, start, end, label="", connectionstyle="arc3"):
    """Draws a directional arrow between coordinates."""
    arrowprops = dict(
        arrowstyle="-|>",
        color='#34495e',
        lw=1.8,
        mutation_scale=15,
        connectionstyle=connectionstyle,
        zorder=4
    )
    ax.annotate('', xy=end, xytext=start, arrowprops=arrowprops)
    if label:
        mid_x = (start[0] + end[0]) / 2
        mid_y = (start[1] + end[1]) / 2 + 0.15
        ax.text(mid_x, mid_y, label, ha='center', va='center', fontsize=7.5, weight='bold', color='#7f8c8d', zorder=5)

def draw_actor(ax, x, y, name):
    """Draws a stick figure actor for Use Case diagrams."""
    # Head
    circle = patches.Circle((x, y + 0.4), 0.15, facecolor='#ffffff', edgecolor='#2c3e50', lw=2, zorder=3)
    ax.add_patch(circle)
    # Body
    ax.plot([x, x], [y + 0.25, y - 0.1], color='#2c3e50', lw=2, zorder=3)
    # Arms
    ax.plot([x - 0.2, x + 0.2], [y + 0.15, y + 0.15], color='#2c3e50', lw=2, zorder=3)
    # Legs
    ax.plot([x, x - 0.15], [y - 0.1, y - 0.4], color='#2c3e50', lw=2, zorder=3)
    ax.plot([x, x + 0.15], [y - 0.1, y - 0.4], color='#2c3e50', lw=2, zorder=3)
    # Name
    ax.text(x, y - 0.65, name, ha='center', va='center', fontsize=9.5, weight='bold', color='#2c3e50', zorder=3)
```

## 3. Generate MTech Thesis MS Word Document
Now we write the Python script that compiles the MTech Thesis write-up using the `python-docx` library.

```python
from docx import Document
from docx.shared import Inches, Pt, RGBColor
from docx.enum.text import WD_ALIGN_PARAGRAPH
from docx.enum.table import WD_TABLE_ALIGNMENT
from docx.oxml import parse_xml
from docx.oxml.ns import nsdecls

def create_thesis_document():
    doc = Document()
    
    # Page margins
    for section in doc.sections:
        section.top_margin = Inches(1.0)
        section.bottom_margin = Inches(1.0)
        section.left_margin = Inches(1.0)
        section.right_margin = Inches(1.0)
        
    # ── Style Helper Functions ───────────────────────────────────────────────
    def set_font(run, name="Calibri", size=11, bold=False, italic=False, color_rgb=(0,0,0)):
        run.font.name = name
        run.font.size = Pt(size)
        run.bold = bold
        run.italic = italic
        run.font.color.rgb = RGBColor(*color_rgb)
        
    def add_heading_1(text):
        p = doc.add_paragraph()
        p.paragraph_format.space_before = Pt(18)
        p.paragraph_format.space_after = Pt(6)
        p.paragraph_format.keep_with_next = True
        run = p.add_run(text)
        set_font(run, "Arial", 16, bold=True, color_rgb=(44, 62, 80))
        return p
        
    def add_heading_2(text):
        p = doc.add_paragraph()
        p.paragraph_format.space_before = Pt(12)
        p.paragraph_format.space_after = Pt(4)
        p.paragraph_format.keep_with_next = True
        run = p.add_run(text)
        set_font(run, "Arial", 13, bold=True, color_rgb=(52, 73, 94))
        return p
        
    def add_body_p(text, bold_prefix="", italic=False):
        p = doc.add_paragraph()
        p.paragraph_format.space_after = Pt(6)
        p.paragraph_format.line_spacing = 1.15
        if bold_prefix:
            r_pre = p.add_run(bold_prefix)
            set_font(r_pre, "Calibri", 11, bold=True)
        run = p.add_run(text)
        set_font(run, "Calibri", 11, italic=italic)
        return p

    def add_bullet_p(text):
        p = doc.add_paragraph(style='List Bullet')
        p.paragraph_format.space_after = Pt(3)
        run = p.add_run(text)
        set_font(run, "Calibri", 11)
        return p

    def add_table_styled(headers, rows):
        table = doc.add_table(rows=1, cols=len(headers))
        table.alignment = WD_TABLE_ALIGNMENT.CENTER
        hdr_cells = table.rows[0].cells
        
        for idx, header in enumerate(headers):
            hdr_cells[idx].text = header
            # style cell
            p = hdr_cells[idx].paragraphs[0]
            p.alignment = WD_ALIGN_PARAGRAPH.LEFT
            p.paragraph_format.space_after = Pt(2)
            set_font(p.runs[0], "Arial", 9.5, bold=True, color_rgb=(255,255,255))
            # background shading (dark slate blue)
            shading_elm = parse_xml(f'<w:shd {nsdecls("w")} w:fill="2c3e50"/>')
            hdr_cells[idx]._tc.get_or_add_tcPr().append(shading_elm)
            
        for row_data in rows:
            row_cells = table.add_row().cells
            for idx, text in enumerate(row_data):
                row_cells[idx].text = str(text)
                p = row_cells[idx].paragraphs[0]
                p.paragraph_format.space_after = Pt(2)
                set_font(p.runs[0], "Calibri", 9.5)
        doc.add_paragraph().paragraph_format.space_after = Pt(12)

    # ── Title Section ────────────────────────────────────────────────────────
    title_p = doc.add_paragraph()
    title_p.alignment = WD_ALIGN_PARAGRAPH.CENTER
    title_p.paragraph_format.space_before = Pt(36)
    title_p.paragraph_format.space_after = Pt(12)
    title_run = title_p.add_run("EDUPULSE: AN INTEGRATED ACADEMIC PORTAL AND CRM\nWITH HYBRID RECOMMENDATIONS AND AI SUPPORT")
    set_font(title_run, "Arial", 18, bold=True, color_rgb=(44, 62, 80))

    subtitle_p = doc.add_paragraph()
    subtitle_p.alignment = WD_ALIGN_PARAGRAPH.CENTER
    subtitle_p.paragraph_format.space_after = Pt(36)
    sub_run = subtitle_p.add_run("Technical Documentation & Thesis Write-up for MTech Degree Project")
    set_font(sub_run, "Arial", 12, italic=True, color_rgb=(127, 135, 143))

    # Page break after title
    doc.add_page_break()

    # ── Chapter 1: Abstract ──────────────────────────────────────────────────
    add_heading_1("1. Executive Summary / Abstract")
    add_body_p(
        "Modern higher education systems frequently suffer from administrative fragmentation, "
        "delays in support ticketing, and an inability to deliver customized academic assistance. "
        "This project presents EduPulse, an integrated, full-stack Academic Portal and CRM "
        "solution optimized for universities. Built using modern web technologies—Next.js 16, "
        "React 19, and Tailwind CSS—along with a backend MySQL database facilitated by Prisma ORM, "
        "EduPulse addresses critical system bottlenecks. Crucially, the platform includes a Python "
        "AI Subsystem powered by FastAPI, featuring an SVD (Singular Value Decomposition) and XGBoost "
        "hybrid recommendation model alongside an LLM-driven intelligent support chatbot. "
        "The resultant system delivers personalized study resources to students based on collaborative "
        "interactions, provides automated real-time database queries and web searches to resolve academic support ticket queries, "
        "and streamlines administrative workflows for campus registrars."
    )

    # ── Chapter 2: Introduction ──────────────────────────────────────────────
    add_heading_1("2. Introduction")
    add_body_p(
        "Higher education environments manage massive student cohorts and extensive resource repositories. "
        "Students need direct access to study materials, clearance statuses, and technical query support. "
        "Administrators require comprehensive dashboards tracking finance and medical clearances, campaign broadcasters, "
        "and instant support ticket desks. Traditional systems lack the intelligence to bridge resources and students "
        "dynamically, leading to poor resource utilization and high support overhead. "
        "The objectives of the EduPulse system are to:"
    )
    add_bullet_p("Provide a responsive academic workspace for student resource browsing and real-time support tickets.")
    add_bullet_p("Empower administrative staff with an educational CRM tracking tuition clearance and campaign dispatch.")
    add_bullet_p("Introduce hybrid machine learning (SVD + XGBoost) to predict resource relevance per student profile.")
    add_bullet_p("Integrate LLMs with real-time database extraction and web search to support ticket automation.")

    # ── Chapter 3: System Architecture ───────────────────────────────────────
    add_heading_1("3. System Architecture & Tech Stack")
    add_body_p(
        "EduPulse employs a decoupled, multi-tier system architecture partitioned into three distinct operational layers:"
    )
    add_body_p(
        "Responsible for rendering the student dashboard and admin interface. Next.js 16 utilizes the App Router "
        "architecture to ensure optimal page loads and caching, while Tailwind CSS v4 provides responsive grids. "
        "State propagation is coordinated through React context Providers.",
        bold_prefix="Presentation Tier: "
    )
    add_body_p(
        "Node.js API routes under the Next.js server handle endpoint requests, secure JWT (jose) verification "
        "contained in HTTP-only cookies, and implement Server-Sent Events (SSE) to broadcast tickets and campaign state changes.",
        bold_prefix="Application / API Tier: "
    )
    add_body_p(
        "Composed of a MySQL database accessed via Prisma ORM for relational CRUD, and a FastAPI (Python) server "
        "hosting machine learning models and LLM wrappers to fulfill real-time recommendation and chatbot requests.",
        bold_prefix="Data & Model Integration Tier: "
    )

    # ── Chapter 4: Database Design ───────────────────────────────────────────
    add_heading_1("4. Database Design & Data Dictionary")
    add_body_p(
        "The application leverages a relational MySQL schema. Below is the data dictionary of the primary tables:"
    )
    
    add_heading_2("Table: Student")
    add_table_styled(
        ["Column Name", "Type", "Constraints", "Description"],
        [
            ["id", "String", "PK, UUID", "Unique primary identifier for the student record."],
            ["name", "String", "Required", "Full legal name of the student."],
            ["matricule", "String", "Unique", "University matriculation number used for credentials."],
            ["instEmail", "String", "Required", "Institutional email address."],
            ["personalEmail", "String", "Required", "Personal contact email address."],
            ["faculty / dept", "String", "Required", "Target faculty and department names."],
            ["level", "Int", "Required", "Academic year (e.g. 100, 200, 300, 400)."],
            ["tuitionStatus", "String", "Required", "Status representing financial clearance ('Paid', 'Overdue')."],
            ["medicalStatus", "String", "Required", "Status representing clinical clearance ('Cleared', 'Pending')."],
        ]
    )
    
    add_heading_2("Table: Resource")
    add_table_styled(
        ["Column Name", "Type", "Constraints", "Description"],
        [
            ["id", "String", "PK, UUID", "Unique identifier for study material."],
            ["title / url", "String", "Required", "Resource title and path to the document/video file."],
            ["type", "String", "Required", "Material categorisation ('pastpaper', 'notes', 'youtube')."],
            ["faculty / dept", "String", "Required", "Filters matching target audience academic departments."],
            ["level", "Int", "Required", "Target academic level for recommendation."]
        ]
    )
    
    add_heading_2("Table: StudentInteraction")
    add_table_styled(
        ["Column Name", "Type", "Constraints", "Description"],
        [
            ["id", "String", "PK, UUID", "Unique primary identifier for the interaction."],
            ["studentId", "String", "FK -> Student.id", "Reference to the student performing the action."],
            ["resourceId", "String", "FK -> Resource.id", "Reference to the target academic resource."],
            ["action", "String", "Required", "Type of action ('view', 'click', 'download', 'complete')."],
            ["duration", "Int?", "Optional", "Time spent reading/watching in seconds."]
        ]
    )

    # ── Chapter 5: Recommendation Engine ──────────────────────────────────────
    add_heading_1("5. AI Recommendation Engine")
    add_body_p(
        "To address information overload and enhance resource discovery, the FastAPI server exposes "
        "a `/recommend` endpoint driven by a hybrid collaborative-filtering model. The scoring pipeline consists of:"
    )
    add_body_p(
        "Singular Value Decomposition (SVD) collaborative-filtering models, trained on interaction "
        "histories, compute latent factor embeddings that score resource suitability for a student ID.",
        bold_prefix="Collaborative Model (SVD): "
    )
    add_body_p(
        "XGBoost is used to rank resources by taking student attributes (level, faculty, department) "
        "and resource features, outputting the probability that a student will engage with the resource. "
        "The final model features are encoded using categorical LabelEncoders saved in joblib format.",
        bold_prefix="Content/Demographic Model (XGBoost): "
    )
    add_body_p(
        "If API credentials are present, a one-sentence descriptive justification "
        "explaining why these particular resources are suitable is generated via GPT-4o-mini, providing "
        "context to the student UI.",
        bold_prefix="LLM Explainer: "
    )

    # ── Chapter 6: Chatbot ───────────────────────────────────────────────────
    add_heading_1("6. Intelligent Chatbot Subsystem")
    add_body_p(
        "The chatbot is designed to handle support requests. The `/chat` endpoint accepts "
        "the user message and a message history list, execution flows as follows:"
    )
    add_bullet_p("Extraction of database context: The server executes SQL queries to extract overall registration numbers, tuition balance metrics, pending clearances, and open support tickets.")
    add_bullet_p("Web Search Augmentation: A keyword routing module inspects incoming queries. If search keywords are matched (e.g., 'latest', 'research', 'recommend'), the DuckDuckGo Search API is triggered to pull auxiliary facts.")
    add_bullet_p("Prompt Synthesis: A system instruction prompt is built, injecting the database stats, search context, and student metadata.")
    add_bullet_p("SSE Streaming Response: The unified payload is forwarded to OpenRouter. Tokens generated by gpt-4o-mini are asynchronously piped back using FastAPI's StreamingResponse as SSE chunks.")

    # ── Chapter 7: Conclusion ────────────────────────────────────────────────
    add_heading_1("7. Conclusion & Future Scope")
    add_body_p(
        "EduPulse represents a cohesive solution integrating a full-stack academic CRM portal "
        "with machine learning and large language models. The hybrid recommender ensures students "
        "find relevant study guides efficiently. The database-aware support chatbot reduces "
        "administrative support ticket workloads. Future iterations will include automated model retraining "
        "pipelines triggered by interaction thresholds, support for multi-modal ticketing attachments, "
        "and deeper integration of predictive analytics on student retention and academic performance."
    )
    
    # Save
    output_path = os.path.join(docs_dir, 'EduPulse_MTech_Thesis_Writeup.docx')
    doc.save(output_path)
    print(f"Saved: {output_path}")

if __name__ == '__main__':
    create_thesis_document()
```
